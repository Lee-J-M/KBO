```python
import pandas as pd
import numpy 
import tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(40,input_dim=4,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(25,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
```

```python
#18년도 데이터로 FITTING
df_18=pd.read_csv("팀투수_2018.csv",parse_dates=['GDAY_DS'])

df_18["T_avg_ER"]=(df_18['ER']*9)/df_18['INN2']
df_18["WHIP"]=(df_18['HIT']+df_18['HR']+df_18['BB'])/df_18['INN2']
del df_18['ER']
del df_18['INN2']

df_18_SS=df_18[df_18['T_ID']=='SS']
df_18_SS=df_18_SS.set_index('GDAY_DS')

#df_18_HT

X_18_SS=df_18_SS[['HIT','WHIP','PA','HR']]
Y_18_SS=df_18_SS['T_avg_ER']

Y_18_SS_sh=Y_18_SS.shift(-29)
#Y_OB_19_sh

model.fit(X_18_SS[:-29],Y_18_SS_sh[:-29],epochs=40,batch_size=10)
```

```python
#예측값 구하기
test_20=pd.read_csv('2020투수기록.csv')

test_20_SS=test_20[test_20['T_ID']=='SS']

test_input_SS=test_20_SS[['HIT','WHIP','PA','HR']]

SS_prediction_20=model.predict(test_input_SS[-29:]).flatten()
```
