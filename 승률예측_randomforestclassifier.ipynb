```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import tensorflow as tf
 
# Read Input Data
data = pd.read_csv("팀투수_통합.csv")


# 이름만 바꿈
OB = data[data['T_ID']=='OB']
HH = data[data['T_ID']=='HH']
data = data[data['T_ID']=='HT']
KT = data[data['T_ID']=='KT']
LG = data[data['T_ID']=='LG']
LT = data[data['T_ID']=='LT']
NC = data[data['T_ID']=='NC']
SK = data[data['T_ID']=='SK']
SS = data[data['T_ID']=='SS']
WO = data[data['T_ID']=='WO']

# data = data[(data['WLS']!='D')]

print(len(data))
data.info()
```

```python
# (승,패,무) 숫자로 바꿔주기
data.WLS = data.WLS.replace('W',1)
data.WLS = data.WLS.replace('L',-1)
data.WLS = data.WLS.replace('D',0)
data.WLS = data.WLS.astype(int)
# 
x_data = np.array(data.iloc[:-30,[9,11,10,12,13,16,21,24,29,30]])
y_data = np.array(data.iloc[30:,7:8])
xx_data = np.array(data.iloc[-20:,[9,11,10,12,13,16,21,24,29,30]])



print(x_data.shape)
print(y_data.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,random_state=66,test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=3, n_estimators=10, max_depth=4)

model
```

```python
# 모델 훈련
model.fit(x_train,y_train)
print("훈련 세트 정확도: {:.3f}".format(model.score(x_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(model.score(x_test, y_test)))
```

```python
# 예측 된 값 보기
prediction = model.predict(xx_data)

prediction
```

```python
#무승부 데이터로 인해 승률예측이 잘 안돼서 무승부를 처리하기 위한 부분
f = open('잔여경기.csv','r')
rdr = csv.reader(f)
lines = []
for i in rdr:
    lines.append(i)
lines 
del lines[0]
print(len(lines))
lines
```

```python
# 최근 경기 (승 무 패) - 무승부일때 최근전적으로 판별
a = [('NC', 9,0,1),('키움',5,0,5),('KT',7,0,3),('LG',5,0,5),('두산',3,0,7),('KIA',4,0,6),
('롯데',4,0,6),('삼성',3,0,7),('SK',3,0,7),('한화',7,0,3)]
# 팀 별 승률(승 패) 
win_rate = [['NC',0,0],['키움',0,0],['KT',0,0],['LG',0,0],['두산',0,0],['KIA',0,0],['롯데',0,0],['삼성',0,0],['SK',0,0],['한화',0,0]]
```

```python
# 무승부일때 최근전적 10개로 승패 판별
for i in range(0,93):
    if lines[i*2][2] == lines[i*2+1][2]:
        name = lines[i*2][0]
        name2 = lines[i*2+1][0]
        one=0
        two=0
        for j in a:
            if j[0] == name:
                one = int(j[1])-int(j[3])
            elif j[0] ==name2:
                two = int(j[1])-int(j[3])
        if one>two:
            lines[i*2][2]='1'
            lines[i*2+1][2]='-1'
        elif one<two:
            lines[i*2][2]='-1'
            lines[i*2+1][2]='1'
        else:
            lines[i*2][2]= '0'
            lines[i*2+1][2]='0'
            
            
# 승률을 계산함
for i in lines:
    for j in win_rate:
        if j[0] == i[0]:
            if i[2] =='1':
                j[1] = j[1]+1
            elif i[2] =='-1':
                j[2] = j[2]+1
            else :
                pass
            if i[1]=='0':
                rate = 0
            elif i[2]=='0':
                rate = 1
            else :
                rate = j[1]/(j[1]+j[2])
            i[3] = rate
                
print(win_rate)
```

```python
# 무승부 문제를 처리 한 잔여경기에 대한 예측결과를 저장
f = open('잔여경기10.csv','w') #원본을 훼손할 위험이 있으니 다른 파일에 저장하는 것을 추천합니다.
wr = csv.writer(f)
for i in lines:
    wr.writerow(i)   
f.close()
```
